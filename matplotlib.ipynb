Q. What is Matplotlib library in python

 --> Matplotlib is a popular data visualization library in Python. 
 It's often used for creating static, interactive, and animated visualizations in Python

Q. some of the major pros of matplotlib are:

--> 

    Generally easy to get started for simple plots

    support for custom labels and text 

    great control of every element in a figure

    High-quality output in many formats 

    very customizable in general


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Simple Examples
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

Scatter Plot

```python
# Plotting using matplotlib
# plt scatter

plt.figure(figsize=(7,5))
plt.scatter(x,y,c = "g",marker="*")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Graph in 2D")
plt.show()
# plt.savefig("test.png")
```

Line Plot

```python
y = x*x
```

```python
# plt plot
plt.figure(figsize=(7,5))
plt.plot(x,y,c = "k",marker = "o",linestyle = "--",linewidth = 2,markersize = 10)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line plot")
plt.show()
```

Subplot

```python
# Creating Subplot

plt.subplot(2,2,1) #(2 rows,2 columns,1 Position)
plt.plot(x,y,"r--")
plt.subplot(2,2,2)
plt.plot(x,y,"b*")
plt.subplot(2,2,3)
plt.plot(x,y,"go--")
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title("matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y,"ko--")
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")

# plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,10))
# subplot()
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2,1,1)

# Make the first plot
plt.plot(x,y_sin,color = "red",linestyle = "-.",linewidth = 2)
plt.title("Sine")

# Set up the second such subplot as active and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,color = "green",linestyle = "-.",linewidth = 2)
plt.title("Cosine")

# Show the figure
plt.show()
```

Bar Plot

```python
# Bar plot

plt.figure(figsize=(10,6))

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color = "g")
plt.title("Bar graph")
plt.ylabel("Y axis")
plt.xlabel("X axis")

plt.show()
```

BoxPlot using Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

# Rectangular box plot
plt.boxplot(data,vert = True,patch_artist=True);
# plt.show()
```

Pie Chart

```python
# Data to plot

labels = "Python","C++","JavaScript","Java"
Sizes = [85,20,49,52]
colors = ["gold","Yellowgreen","green","lightblue"]
explode = (0.1,0,0,0) # explode 1st slice

# plot

plt.pie(Sizes,explode=explode,labels=labels,colors=colors,autopct = "%1.1f%%", shadow = True)

plt.axis = ("equal")
plt.show()
```

```python
plt.pie(Sizes,explode=explode,labels=labels,colors=colors,autopct = "%1.1f%%", shadow = True)

cr = plt.Circle(xy=(0,0),radius=0.5,facecolor="w")
plt.gca().add_artist(cr)

plt.axis = ("equal")
plt.show()
```

Stem Plots

```python
x = [1,2,3,4,5,6]
y = [2,2,5,7,3,8]
plt.stem(x,y,linefmt = ":",markerfmt="r+",basefmt="g",label="Python")

plt.legend()
plt.show()
```

Stack Plot(Area Plot)

```python
k = [1,2,3,4,5]
area = [1,2,3,6,5]
area1 = [7,2,9,6,5]
area2 = [3,8,3,7,4]

lable = ["area","area1","area2"]
color = ["lightgreen","yellowgreen","brown"]
plt.stackplot(k,area,area1,area2,labels=lable,colors=color)

plt.title("Area plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc = "best")

plt.show()
```

Step Plot

```python
v = [1,2,3,4,5]
w = [2,2,3,5,4]

plt.step(v,w,color = "blue",marker = "o",label = "Python")

plt.title("Step plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc = "best")
plt.grid()
plt.show()
```

```python
plt.plot(v,w,marker = "o",color = "blue")
plt.show()
```

Fill Between Plot

```python
p = np.arange(1,6)
q = np.arange(1,11,2)
print(p.dtype)
print(q.dtype)

plt.plot(p,q,color="r")

x = np.arange(1,6)
y = np.arange(1,11,2)
plt.fill_between(x,y,color = "g", where=(x>=2) & (x<=4))

plt.title("python")
plt.xlabel("year")
plt.ylabel("Popularity")
plt.show()
```

Text in Plot

```python
x = ["2019","2020","2021","2022","2023"]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("Growth per year",fontsize = 20)
plt.xlabel("year")
plt.ylabel("revanue in B$")

plt.text("2023",6,"Python",fontsize = 15,bbox = {"facecolor":"red"})
plt.text("2021",5,"c++",fontsize = 15,bbox = {"facecolor":"red"})
plt.text("2019",3,"SQL",fontsize = 15,bbox = {"facecolor":"red"})

plt.annotate("java",xy=("2022",2),xytext=(3,4),arrowprops=dict(facecolor = "green"))
plt.annotate("c",xy=("2020",1),xytext=(1,4),arrowprops=dict(facecolor = "green"))
```
