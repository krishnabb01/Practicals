Prctice Problem on IRIS Dataset

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

**Correlation with Heatmap**

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix(table) between Two discreate dimensions or events types. It is very important Feature Selecton

```python
k = df[["total_bill", "tip", "size"]].corr()
k
```

```python
sns.heatmap(k)
```

**JoinPlot**

A join plot allows to study the relationship between 2 numeric variables. The central chart display their corrleation. It is usually a scatterplot a hexbin plot. a 2D histogram or 2D density plot

**Univarite Analysis**

```python
sns.jointplot(x="tip",y="total_bill",data=df, kind="hex")
```

```python
sns.jointplot(x = "tip", y = "total_bill",data = df,kind = "reg")
```

**Pair plot**

A "pairs plot" is also known as scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pair plots are just elaborations on this, showing all variables paired with all the other varibles

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue = "sex")
```

**Dist plot**

Dist plot helps us to check the distribution of the columns feature

```python
sns.displot(df["tip"],kde = True, bins = 10)
```

```python
sns.displot(df["tip"],kde = False, bins = 10)
```

**Categorical Plots**

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

- boxplot
- violinplot
- countplot
- bar plot

```python
# Count Plot

sns.countplot(x= "day",data = df,palette="gist_ncar")
```

```python
sns.countplot(y = "sex",data = df,palette="gist_ncar")
```

```python
# Bar plot
sns.barplot(x = "total_bill",y = "smoker",data = df,palette="gist_ncar")
```

```python
# Bar plot
sns.barplot(x = "sex",y = "total_bill",data = df,palette="gist_ncar")
```

**Box Plot**

A box and whisker plot(sometimes called a boxenplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x = "sex", y = "total_bill",data = df,palette="rainbow")
```

```python
sns.boxplot(x = "day", y = "total_bill",data = df,palette="rainbow")
```

```python
# Categorize my data based on some other categories

sns.boxplot(x = "day", y = "total_bill",hue="sex",data = df,palette="rainbow")
```

**Violin Plot**

Violin plot helps us to see both the distribution of data in terms of kernel density estimation and box plot

```python
sns.violinplot(x = "total_bill",y = "day",data = df,palette="rainbow")
```

**Scatter plot**

```python
sns.scatterplot(x = "total_bill", y = "tip",data = df,hue = "sex",palette="rainbow",size = "sex",sizes=(40,40),style="sex")
```

**Line Plot**

```python
sns.lineplot(x = "day", y = "size", data = df)
```

**Bar plot**

```python
sns.barplot(x = "day", y = "size", data = df,hue="sex",hue_order=["Female","Male"],palette="icefire",errcolor="red",capsize=0.2)
```

**Heat Map**

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = df.pivot_table(values="total_bill", index="day", columns="time")

c = {"fontsize": 10, "color": "black"}

# Create the heatmap
sns.heatmap(data,cmap="Blues",annot=True,annot_kws=c,linewidths=1,linecolor="r",cbar=True,
            xticklabels=True,  # Adjust according to your needsyticklabels=True   # Adjust according to your needs
)

# Set axis labels
plt.xlabel("Time")
plt.ylabel("Day")

# Display the plot
plt.show()
```

**Strip Plot**

```python
sns.stripplot(x = "day",y ="total_bill",data=df,palette="rainbow",hue="sex",linewidth=0.8,size=10,alpha = 0.5)
plt.show()
```

**Cat Plot**

```python
sns.catplot(x = "day",y = "total_bill",data=df,hue="sex",palette="gist_ncar")
plt.show()
```

**"Multipal Plot" Facetgrid**

```python
kb = sns.FacetGrid(df,col="day",hue="sex",palette="gist_ncar")
kb.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```
